# TROUVER UN FILM AVEC la reference

Avec l'id de référence

```python
def find_k_films_bert_similarity_with_reference(model,dataset, reference_line, k=5):
  reference_plot = dataset.loc[reference_line, 'Plot']

  # Encodage de la référence
  reference_embedding = model.encode([reference_plot])

  #Maintenant on doit essayer de trouver les k films les plus proches selon les embeddings
  # Liste pour stocker les informations des k films les plus similaires
  top_k_movies_info = []

  for line in dataset.index:
        if line != reference_line:
            # Encodage du plot du film
            tokenized_plot = dataset.loc[line, 'Plot']
            plot_embedding = model.encode([tokenized_plot])


            # Calcul de la similarité cosine entre la référence et le plot du film
            similarity_score = cosine_similarity(reference_embedding, plot_embedding)[0][0]

            movie_info = {
                'Title': dataset.loc[line, 'Title'],
                'Director': dataset.loc[line, 'Director'],
                'Release Year': dataset.loc[line, 'Release Year'],
                'Similarity Score': similarity_score
            }

            top_k_movies_info.append(movie_info)

  # Trier la liste par scores de similarité décroissants
  top_k_movies_info = sorted(top_k_movies_info, key=lambda x: x['Similarity Score'], reverse=True)[:k]

  return top_k_movies_info
```

```python
western_movies = dataset[dataset['Genre'] == 'western']
western_movies_2008 = western_movies[(western_movies['Release Year'] >= 2008) & (western_movies['Release Year'] <= 2020)]
```

```python
reference_movie_id = 16253 # FIlm DJANGO UNCHAINED
top_k_similar_movies = find_k_films_bert_similarity_with_reference(model,western_movies_2008,reference_movie_id,k=3)
print(top_k_similar_movies)
```

AVEC LA DESCRITPION UTILISATEUR

```python
def find_k_films_bert_similarity_with_user_desc(model, dataset, plot, k=5):
    # Encodage du plot du film référence
    reference_embedding = model.encode([plot])

    # Maintenant on doit essayer de trouver les k films les plus proches selon les embeddings
    # Liste pour stocker les informations des k films les plus similaires
    top_k_movies_info = []

    for line in dataset.index:
        # Récupération des embeddings des plots
        plot_embedding = dataset.loc[line, 'Plot_Embedding']
        plot_embedding = np.fromstring(plot_embedding[1:-1], sep=' ')

        # Calcul de la similarité cosine entre la référence et le plot du film
        similarity_score = cosine_similarity(reference_embedding, plot_embedding.reshape(1, -1))[0][0]

        # Stockage des informations des films et du score de similarité
        movie_info = {
            'Title': dataset.loc[line, 'Title'],
            'Genre': dataset.loc[line, 'Genre'],
            'Director': dataset.loc[line, 'Director'],
            'Release Year': dataset.loc[line, 'Release Year'],
            'Similarity Score': similarity_score
        }

        # Ajout des films dans la liste
        top_k_movies_info.append(movie_info)

    # Tri de la liste par scores de similarité décroissants et récupération des k films avec le score le plus élevé
    top_k_movies_info = sorted(top_k_movies_info, key=lambda x: x['Similarity Score'], reverse=True)[:k]

    # Affichage des films dans la console
    # Pour le debugage
    # print(top_k_movies_info)

    # Renvoi des k films à recommander
    return top_k_movies_info
```

```python
films_harry_potter = dataset[dataset['Title'].str.contains('harry potter', case=False)]
```

```python
user_description = "Hey i would like to watch a Harry Potter Movie, where Dobby appears"
recommended_films = find_k_films_bert_similarity_with_user_desc(model,films_harry_potter,user_description,k=3)
print(recommended_films)
```
